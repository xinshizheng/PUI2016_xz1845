```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
print os.environ.get('DFDATA')
```

```python
data_location = os.environ.get('DFDATA') + '/zkky-n5j3/1414246141/zkky-n5j3'
data = pd.read_csv(data_location)
data.head()
```

```python
data_2 = data.drop(data.columns.values.tolist()[2:21], axis = 1)
```

```python
data_2
```

```python
df_date = pd.to_datetime(data_2[data_2.columns.values.tolist()[0]])
```

```python
df_elev = data_2[[data_2.columns.values.tolist()[1]]]
```

```python
df_date = pd.DataFrame(df_date)
df = pd.concat([df_date, df_elev], axis = 1)
```

```python
df.plot(x = df.columns.values.tolist()[0], y = df.columns.values.tolist()[1], rot=90)
plt.legend(loc = 4)
plt.ylabel('Elevation')
plt.title('Historical Elevation Change')
plt.show()
```
