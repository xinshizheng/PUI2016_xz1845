```python
# Import packages
import os
import pandas as pd
import geopandas as gpd
from fiona.crs import from_epsg
from shapely.geometry import Point
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Download data
os.system("mkdir " + os.getenv("PUIDATA")+"/CensusBlock")
os.system('curl -O http://www1.nyc.gov/assets/planning/download/zip/data-maps/open-data/nycb2010_16d.zip')
os.system("mv " + "nycb2010_16d.zip " + os.getenv("PUIDATA") + "/CensusBlock")
os.system("unzip " + os.getenv("PUIDATA") + "/CensusBlock/" + "nycb2010_16d.zip -d " + os.getenv("PUIDATA") + "/CensusBlock")
os.system("rm " + os.getenv("PUIDATA") + "/CensusBlock/nycb2010_16d.zip")
```

```python
# Load census block data
nycb_path = os.getenv("PUIDATA") + "/CensusBlock/nycb2010_16d/nycb2010.shp"
nycb_shape = gpd.GeoDataFrame.from_file(nycb_path)
```

```python
# Change the coordination system
nycb_shape.crs = from_epsg(2263)
nycb_latlon = nycb_shape.to_crs(epsg=4326)
```

```python
# Subset Brooklyn census blocks
bkcb_latlon = nycb_latlon[nycb_latlon['BoroName'] == 'Brooklyn']
```

```python
bkcb_latlon.head()
```

```python
# CUSP location
cusp_lat = 40.693158
cusp_lon = -73.986186
```

```python
# Make the CUSP a point feature
cusp = Point(cusp_lon, cusp_lat)
```

```python
# Find the census block that CUSP is in
cusp_cb = bkcb_latlon[bkcb_latlon.geometry.contains(cusp)]
```

```python
# Plot every 10 census blocks with CUSP and the census block it is in highlighted
ax = bkcb_latlon[::10].plot(figsize=(10,10), linewidth=0.1, color='grey', alpha=0.1)
cusp_cb.plot(ax=ax, alpha=1, color='red', linewidth=1)
plt.scatter(cusp.x, cusp.y, c='green', s=60, alpha=0.5)
ax.set_axis_off()
```
